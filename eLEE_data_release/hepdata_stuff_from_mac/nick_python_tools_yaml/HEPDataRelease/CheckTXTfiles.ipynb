```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
mc = np.loadtxt('nue_eventlist.txt')
binedges = np.loadtxt('nue_1e1p_bins.txt')
bc = 0.5*(binedges[:-1]+binedges[1:])
data = np.loadtxt('nue_1e1p_data.txt')
```

```python
n,_,_ = plt.hist(mc[:,0],bins=binedges,weights=mc[:,-1])
plt.plot(bc,data)
plt.show()
```

```python
plt.hist2d(mc[:,1],mc[:,0],bins=(13,10),range=((200,1500),(200,1200)),weights=mc[:,-1],norm=LogNorm())
plt.colorbar()
plt.plot([0,2000],[0,2000],color='black')
plt.ylabel('EnuReco')
plt.xlabel('EnuTrue')
plt.show()
```
